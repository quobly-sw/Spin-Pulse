# Generation of time traces and notion of coherent time

In our model, a qubit evolves via a discretized time-dependent Hamiltonian $h(t) = \epsilon(t) Z$, 
with $t=t_s,\dots,N t_s$, where $t_s$ is the sampling time and acts as our unit of time in our code suite (i.e., $t_s=1$ in what follows.), and $N$ is the length of the time trace.

We have three types of noise: coherent, white, and pink noise implemented

For a qubit initialized in $\ket{0}+\ket{1}$, the observable $X$ will decay in time due to the noise Hamiltonian $h(t)$.

Repeating this experiment multiple times, we will obtain a Ramsey signal $\braket{X(t)}=\frac{1}{2}(1+C)$
with the Ramsey contrast $C=\mathbb{E}[e^{-i\sum_{t'\le t}\epsilon(t')}]$, and $\mathbb{E}$ denotes an ensemble average over multiple random realizations of the time trace.



## Coherent noise

In the case of coherent noise, we have a time-independent signal $\epsilon(t)=\epsilon$ where $\epsilon$ is chosen from a random normal distribution of standard deviation $\sigma$.

This leads to
$$
C=\mathbb{E}[\exp(-it \epsilon)]=\exp(-t^2\sigma^2/2)=\exp(-t^2/T_2^2)
$$
with $T_2=\sqrt{2}/\sigma$. Note: the first equality corresponds to the expression of the characteristic function of the normal distribution.

### White noise

In the case of white noise, we have a time-dependent signal where $\epsilon(t)$ where $\epsilon(t)$ is chosen from a random normal distribution of standard deviation $\sigma$.

$$C=\mathbb{E}[\exp(-i\epsilon(1))]^t=\exp(-t\sigma^2/2)=\exp(-t/T_2)$$

with $T_2=2/\sigma^2$. Remember that the code is expressed in the unit of discretization time $t_s$, so we have $T_2=2t_s/\sigma^2$ in SI units.

```python
time_trace = WhiteNoiseTimeTrace(T2, duration, 1)
time_trace.plot(n_max=2000)
plt.savefig("../paper/fig3b.pdf", bbox_inches="tight")
```

```python
time_trace.plot_ramsey_contrast(ramsey_duration)
# plt.savefig("../paper/fig3e.pdf", bbox_inches="tight")
```

### Pink noise

We consider time traces of the form $\omega(t)=2\pi \sqrt{S_0} g(t)$, where $g(t)$ is a normalized time trace associated with a Power-Spectral Density $S_g(f)=1/f$, where $f$ is the discretized frequency associated with the time vector $t$ introduced above, and $S_0$ is the spectral intensity (which has dimension of the square frequency, and is thus written in our code in units of $t_s^{-2}$)

Following [Yoneda, et al, 2018](https://doi.org/10.1038/s41565-017-0014-x.), and refs therein, the Ramsey contrast can be approximated as follows

$$
C\approx \exp(-t^2/T_2^\star)
$$

with a \emph{time-dependent} value of the coherence time

$$
T_2^\star(t) = \frac{1}{2\pi\sqrt{S_0\log(\frac{N}{t})}}
$$. 

In the short-time limit, we can assume $T_2^\star(t)=T_2^\star(1)=T_2^\star$. This  gives us a possibility to choose the value of $S_0=1/(4\pi^2\log(N)[T_2^\star]^2)$, for a certain time trace duration $N$ to achieve a certain value of $T_2^\star$

Note that the value of $N$ controls a low-frequency cutoff $f_{\min}=1/N$ in the spectral properties of the time trace. By increasing $N$, the signal will have more low-frequency components, and the signal will be more and more closer to a Gaussian.

```python
segment_duration = 2**15
time_trace = PinkNoiseTimeTrace(T2, duration, segment_duration)
# time_trace.plot(n_max=2000)
# plt.savefig('../paper/fig3c.pdf',bbox_inches="tight")
plt.figure(figsize=(20, 4))
plt.plot(time_trace.values)
```
